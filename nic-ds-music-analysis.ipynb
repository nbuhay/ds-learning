```python
# determine qualitative vs quantitative data
# modified from -https://towardsdatascience.com/machine-learning-with-python-classification-complete-tutorial-d2c99dc524ec
def _check_col_type(df: pd.DataFrame, col: str, max_cat=20) -> bool:
    """ determine if column is quantitative or qualitative """
    if (df[col].dtype == "O") | (df[col].nunique() < max_cat):
        res = "quan"
    else:
        res = "qual"
    return res
```

```python
dic_cols = {col:_check_col_type(df, col) for col in df.columns}
dic_cols
```

```python
# check for missing values

heatmap = df.isnull()
heatmap
```

```python
total = 0
for k,v in dic_cols.items():
    if v == "quan":
        heatmap[k] = heatmap[k].apply(lambda x: 0.5 if x is False else 1)
    else:
        heatmap[k] = heatmap[k].apply(lambda x: 0 if x is False else 1)

sns.heatmap(heatmap, cbar=False)
```
